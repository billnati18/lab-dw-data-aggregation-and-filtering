# Lab | Data Aggregation and Filtering

In this challenge, we will continue to work with customer data from an insurance company. We will use the dataset called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by first performing data cleaning, formatting, and structuring.

```python
import pandas as pd
url='https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv'
df = pd.read_csv(url)
df
```

1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

```python
filter_claim = df['Total Claim Amount']>1000
filter_marketing = df['Response']=='Yes'
df.loc[filter_claim&filter_marketing]
```

2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
pd.pivot_table(df.loc[filter_marketing], index=['Policy Type', 'Gender'], values= 'Total Claim Amount', aggfunc='mean')
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
states_500 = df['State'].value_counts()[df['State'].value_counts()>500].index
filter_state = df['State'].isin(states_500)

df[filter_state].groupby("State").size()
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python

df.groupby(['Education', 'Gender'])[['Customer Lifetime Value']].agg(['max', 'min', 'median']).round(2)
```

5. The marketing team wants to analyze the number of policies sold by state and month. Present the data in a table where the months are arranged as columns and the states are arranged as rows.

```python
df.columns
```

```python
from datetime import timedelta
timedelta()
```

```python
df['days_since_policy_inception'] = (df["Months Since Policy Inception"]*365/12).apply(pd.Timedelta, unit='D')
```

```python
df['policy_creation'] = pd.to_datetime(df['Effective To Date'])-df['days_since_policy_inception']
```

```python
df['month'] = df['policy_creation'].dt.month
```

```python
pd.pivot_table(df, index= 'State', columns='month', values= 'Number of Policies')
```

6.  Display a new DataFrame that contains the number of policies sold by month, by state, for the top 3 states with the highest number of policies sold.

*Hint:*
- *To accomplish this, you will first need to group the data by state and month, then count the number of policies sold for each group. Afterwards, you will need to sort the data by the count of policies sold in descending order.*
- *Next, you will select the top 3 states with the highest number of policies sold.*
- *Finally, you will create a new DataFrame that contains the number of policies sold by month for each of the top 3 states.*

```python
top_3_states = df.groupby('State')['Number of Policies'].sum().sort_values(ascending=False).head(3).index
df.loc[df['State'].isin(top_3_states)].groupby(['State', 'month'])['Number of Policies'].sum()
```
